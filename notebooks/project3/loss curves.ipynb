````python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
````

````python
good = np.load('out_succes/losses.npz')['losses']
bad = np.load('out_fail/losses.npz')['losses']
````

```python
lossses[0]: iteration
lossses[1]: dist
lossses[2]: loss
lossses[3]: regular loss
```

````python
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
ax[0].plot(good[:,0], good[:,1], label='Aligned')
ax[0].plot(bad[:,0], bad[:,1], label='Unaligned')
ax[0].set_title('Distance')
ax[0].legend()
ax[1].semilogy(good[:,0], good[:,2], label='Aligned')
ax[1].semilogy(bad[:,0], bad[:,2], label='Unaligned')
ax[1].set_title('Total loss')
ax[1].legend()
ax[2].semilogy(good[:,0], good[:,3], label='Aligned')
ax[2].semilogy(bad[:,0], bad[:,3], label='Unaligned')
ax[2].set_title('Regular loss')
ax[2].legend()
fig.suptitle('Errors with and without face alignment')
fig.savefig('losses.png',dpi=300,bbox_inches='tight')
````
