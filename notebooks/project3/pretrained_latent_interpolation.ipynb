```python
import torch
import numpy as np

import os
from pathlib import Path

from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
torch.cuda.is_available()
```

```python
DATA_PATH = Path('/work3/s194235/GAN/stylegan2-ada-pytorch')
```

```python
example_number = 1
img_name = 'align-IMG_4345'
feature = 'age' # age, smile, yaw
interval = (-30, 30)
```

```python
projected_w = np.load(DATA_PATH / f'projected/{img_name}/projected_w.npz')['w']
projected_feature = np.load(DATA_PATH / f'projected/{feature}/{feature}.npy')
projected_w.shape, np.expand_dims(projected_feature, 0).shape
```

```python
def transform(z, w, m):
    '''
    z: latent vector
    w: feature vector
    m: magnitude of feature
    '''
    return z + w * m


projected_w = np.load(DATA_PATH / f'projected/{img_name}/projected_w.npz')['w']
projected_feature = np.load(DATA_PATH / f'projected/{feature}/{feature}.npy')

for example_number, trade_off in tqdm(enumerate(np.linspace(interval[0], interval[1], 11))):
    trade_off = np.round(trade_off, 1)
    
    interpolated_w = transform(projected_w, np.expand_dims(projected_feature, 0), trade_off)
    os.makedirs(DATA_PATH / f"latent_interpolation/example{example_number}/projected_ws", exist_ok=True)
    np.savez(
        DATA_PATH / f"latent_interpolation/example{example_number}/projected_ws/projected_w_{trade_off}.npz", 
        w=interpolated_w,
    )

for example_number, trade_off in tqdm(enumerate(np.linspace(interval[0], interval[1], 11))):
    trade_off = np.round(trade_off, 1)
    print(f"\n{'-'*100}")
    print(f"Running with trade-off = {trade_off}\n")
    
    !python generate.py --outdir=latent_interpolation/example{example_number}/images --projected-w=latent_interpolation/example{example_number}/projected_ws/projected_w_{trade_off}.npz --network=https://nvlabs-fi-cdn.nvidia.com/stylegan2-ada-pytorch/pretrained/ffhq.pkl
    
```

```python
fig, axs = plt.subplots(1, 11, figsize=(30, 10))
for i, trade_off in enumerate(np.linspace(interval[0], interval[1], 11)):
    trade_off = np.round(trade_off, 1)
    
    img = Image.open(f"latent_interpolation/example{i}/images/proj00.png")
    axs[i].imshow(img)
    axs[i].axis('off')
    axs[i].set_title('$\mathbf{weight:}$ ' + str(trade_off), loc='left', fontsize=15)

plt.tight_layout()
plt.savefig(f"latent_interpolation/interpolated_image_{feature}.png", dpi=300, bbox_inches='tight')
plt.show()
```
